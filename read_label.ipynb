```python
import pandas as pd
import numpy as np
```

```python
def recognise_ints(int_array):
    nests = []
    for s in int_array:
        nests.append(int(s))
    return nests

def recognise_regions(string_array):
    nests = []
    for s in string_array:
        split = s.split('-')
        if len(split) > 1:
            rng = np.arange(int(split[0]),int(split[1])+1, dtype=int)
            nests = list(np.concatenate([nests,rng]).astype('int'))
        elif len(split) ==1:
            print('recognise single int')
    return nests
```

# trim data frame to relevant columns only

```python
col_list = ['Flight N', 'Frames 0', 'Frames 1', 'Frames 2']
df = df[col_list]
df.head(4)
```

# create array of dictionaries for each flight and the frames, where nests have been found

```python
nest_array = []
for index, row in df.iterrows():
    
    flight = row['Flight N']
    n0 = row['Frames 0']
    n1 = row['Frames 1']
    n2 = row['Frames 2']
    
    dic = {'Flight':flight, 'Frames 0': n0 ,'Frames 1': n1, 'Frames 2': n2}
    nest_array.append(dic)
```

```python
nest_array[0]
```

# add to each dictionary an array of all frames, that contain a nest

```python
for flight_dic in nest_array:
    print('Flight:', flight_dic['Flight'])
    new_keys, nests = [], []
    for key,value in flight_dic.items():

        if key != 'Flight':
#             print(key,value)
            if key == 'Frames 0':
                new_keys.append('Nests 0')
            elif key == 'Frames 1':
                new_keys.append('Nests 1')
            elif key == 'Frames 2':
                new_keys.append('Nests 2')
            
            if type(value) == float:
                nests.append([])
                print('\t done with key ', key)
                continue
            str_sp = value.split(',')
    #         print(str_sp)
            if len(str_sp) >1:
    #             print('recognise multiple regions')
    #             nests = list(np.concatenate([nests,recognise_regions(str_sp)]))
    #             print('nests:', recognise_regions(str_sp))
                nests.append(recognise_regions(str_sp))
            elif len(str_sp) == 1:
    #             print(str_sp)
    #             print('recognise single region or single int')
                if len(str_sp[0].split('-')) > 1:
    #                 print('recognise single region')
    #                 print(recognise_regions(str_sp))
    #                 nests = list(np.concatenate([nests,recognise_regions(str_sp)]))
    #                 print('nests:', nests)
                    nests.append(recognise_regions(str_sp))
                elif len(str_sp[0].split('-')) == 1:
    #                 print('recognise single int')
    #                 print(recognise_ints(str_sp))
                    nests.append(recognise_ints(str_sp))
            print('\t done with key ', key)

    for c, key in enumerate(new_keys, 0):
        flight_dic[key] = nests[c]
```
